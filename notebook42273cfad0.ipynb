```python
df['MeanSatisfaction'] = \
    (df['RelationshipSatisfaction']  + df['EnvironmentSatisfaction'] \
     + df['JobSatisfaction'] )/3
df.drop(["RelationshipSatisfaction","EnvironmentSatisfaction",\
        "JobSatisfaction"],axis="columns", inplace=True)
```

```python
object_col = []
for column in df.columns:
    if df[column].dtype == object :
        object_col.append(column)
        print(f"{column} : {df[column].unique()}")
#         print(df[column].value_counts())
        print("====================================")
```

```python
label = LabelEncoder()
df["OverTime"] = label.fit_transform(df.OverTime)
df["Gender"] = label.fit_transform(df.Gender)
```

```python
object_col = []
for column in df.columns:
    if df[column].dtype == object :
        object_col.append(column)
        print(f"{column} : {df[column].unique()}")
#         print(df[column].value_counts())
        print("====================================")
```

```python
df.columns
```

```python
data = pd.get_dummies(df, columns=object_col, drop_first=True, dtype='uint8')
data.head()
```

```python
data.info()
```

```python
y=data.Attrition
x=data.drop("Attrition",axis=1)
print("Before OverSampling, counts of label '1': {}".format(sum(y==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y==0)))
x
```

```python
from imblearn.over_sampling import SMOTE

sm = SMOTE(random_state=2)
x_res, y_res = sm.fit_resample(x,y.ravel())
x_res
```

```python
print("After OverSampling, counts of label '1': {}".format(sum(y_res==1)))
print("After OverSampling, counts of label '0': {} \n".format(sum(y_res==0)))
```

```python
random_state = 42
from sklearn.preprocessing import StandardScaler,MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
# # x_res
x_scaled = scaler.fit_transform(x_res)
# x_scaled
```

```python
x_train, x_test, y_train, y_test = \
    train_test_split(x_scaled, y_res, test_size = 0.3, random_state = 40)
print(len(x_train))
print(len(x_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(random_state=0)
nb=GaussianNB()
knn=KNeighborsClassifier()
```

```python
from sklearn.ensemble import VotingClassifier
final_model=VotingClassifier(estimators= \
                [ ('nb', nb),('knn',knn),("dt",dt)],voting="hard")
final_model.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report
```

```python
def model_eval(model,y_test,y_predict):
    conf_matrix = confusion_matrix(y_test, y_predict)
    accuracy = accuracy_score(y_test, y_predict)
    print("accuracy",accuracy)
    print(classification_report(y_test, y_predict))
```

```python
model = final_model
model.fit(x_train,y_train)
y_predict = model.predict(x_test)
model_eval(model,y_test,y_predict)
```
